```python
#unit 10 qn 1
s_list = ['abc','bcd','bcdefg','abba','cddc','opq']
minimum = len(s_list[0])
smallest_str = s_list[0]
for i in s_list:
    if len(i) < minimum:
        minimum = (len(i))
        smallest_str = i
print("The shortest string:",smallest_str)
```

```python
#Unit 12 qn 1
number = (1,2,5,4,3,2,1,4,7,8,9,9,3,7,3,9)
most_repeated = 1
count = 0
for i in number:
    repeat = number.count(i)
    if repeat > count:
        most_repeated = i
        count = repeat
    elif count == repeat:
        if most_repeated < i:
            most_repeated = i

print("The most frequent number is: ",most_repeated)
```

```python
#Unit 14
#Qn 1
items = {
    "Coffee": 7,
    "Pen": 3,
    "Paper cup": 2,
    "Milk": 1,
    "Coke": 4,
    "Book": 5
}

while True:
    print("Select menu 1)check stock 2)warehousing 3)release 4)exit:", end=" ")
    menu = input()

    if menu == "1":
        item = input("[Check stock] Enter item: ")
        if item in items:
            print("Stock:", items[item])
        else:
            print("Item not found")

    elif menu == "2":
        item = input("[Warehousing] Enter item: ")
        qty = int(input("Enter quantity: "))
        if item in items:
            items[item] += qty
        else:
            items[item] = qty

    elif menu == "3":
        item = input("[Release] Enter item: ")
        qty = int(input("Enter quantity: "))
        if item in items:
            items[item] -= qty
            if items[item] < 0:
                items[item] = 0
        else:
            print("Item not found")

    elif menu == "4":
        print("Program exited.")
        break

    else:
        print("Wrong menu")
```

```python
#Unit 16 qn 1
my_lst = [(1, 2), (4, 5), (4, 2), (3, 1), (9, 4)]
ordinals = {0: "first", 1: "second", 2: "third", 3: "fourth", 4: "fifth"}

ele = input("Enter two integers: ").split()
a,b = int(ele[0]), int(ele[1])

target = (a, b)
reversed_target = (b, a)

if target in my_lst:
    print(f"There is {target} at the {ordinals[my_lst.index(target)]}.")

elif reversed_target in my_lst:
    print(f"There is no {target} but there is {reversed_target} at the {ordinals[my_lst.index(reversed_target)]}.")

else:
    print(f"There is no {target} nor {reversed_target}")
```
