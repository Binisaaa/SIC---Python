```python
#unit 10 qn 1

prime_list = []
for num in range(2,10):
    count = 0
    for n in range(2,num):
        if num % n == 0:
            count += 1
    if count == 0:
         prime_list.append(num)
print("1st element of prime_list:",prime_list[0])
```

```python
#unit 15 qn 1

students_tup = (('211101','David Doe',"010-1234-4500"),('211102','John Smith',"010-2230-6540"),('211103','Jane Carter',"010-3232-7788"))
students_dic = {}
for i in students_tup:
    new_students_dic = {i[0]:[i[1],i[2]]}
    students_dic.update(new_students_dic)

for key,values in students_dic.items():
    print(f"{key}:{values}")
```
