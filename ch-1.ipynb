```python
n = int(input("Enter n: "))
num = 1

for i in range(n):
    row = []
    for j in range(n):
        row.append(num)
        num += 1

    if i % 2 == 1:
        row.reverse()

    for value in row:
        print(value, end=" ")
    print()
```

```python
n = int(input("Enter an integer: "))
temp = n
rev = 0

while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10

if rev == n:
    print(n, "is a palindrome number")
else:
    print(n, "is not a palindrome number")
```

```python
import random

answer = random.randint(1, 100)
count = 0

while True:
    guess = int(input("Enter a number: "))
    count += 1

    if guess > answer:
        print("Lower")
    elif guess < answer:
        print("Higher")
    else:
        print("Congratulations!")
        print("Total try =", count)
        break
```
