```python
#unit 27 qn 1
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def are_anagrams(word1, word2):
    word1 = list(word1.lower())
    word2 = list(word2.lower())
    
    if len(word1) != len(word2):
        return False
    
    return insertion_sort(word1) == insertion_sort(word2)

print(are_anagrams("LISTEN", "SILENT"))
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

def are_anagrams(word1, word2):
    word1 = list(word1.lower())
    word2 = list(word2.lower())
    
    if len(word1) != len(word2):
        return False
    
    return selection_sort(word1) == selection_sort(word2)

print(are_anagrams("LISTEN", "SILENT"))
```

```python
#built-in
def is_anagram(word1, word2):
    word1 = word1.lower()
    word2 = word2.lower()
    
    if len(word1) != len(word2):
        return False    
    return sorted(word1) == sorted(word2)

print(is_anagram("listen", "silent"))
print(is_anagram("anagram", "nagaram"))
print(is_anagram("listen", "silence"))
print(is_anagram("anagram", "anagrams"))
```

```python
# unit 28 qn 1
def merge(list1, list2):
    merged = []
    i = j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1
    merged.extend(list1[i:])
    merged.extend(list2[j:])

    return merged

def multway_merge(lists):
    if len(lists) == 1:
        return lists[0]

    mid = len(lists) // 2
    left = multway_merge(lists[:mid])
    right = multway_merge(lists[mid:])

    return merge(left, right)

N = int(input("Input the number of list: "))
list_of_nums = []
for i in range(N):
    nums = list(map(int, input("Input a list of numbers: ").split()))
    print(nums)
    list_of_nums.append(nums)
sorted_list = multway_merge(list_of_nums)
print("Merged into :", sorted_list)
```

```python
#unit 29 qn 1
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1

def quick_select(arr, low, high, k):
    if low <= high:
        pivot_index = partition(arr, low, high)
        
        if pivot_index == k:
            return arr[pivot_index]
        elif pivot_index > k:
            return quick_select(arr, low, pivot_index - 1, k)
        else:
            return quick_select(arr, pivot_index + 1, high, k)

def find_kth_largest(arr, k):
    n = len(arr)
    target_index = n - k
    return quick_select(arr, 0, n - 1, target_index)

arr = [7, 10, 4, 3, 20, 15]
k = 3
print(find_kth_largest(arr, k))
```
