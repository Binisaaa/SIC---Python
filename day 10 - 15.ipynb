```python
# 4. Write a function common_elements(set1, set2) that takes two sets and returns their intersection.

def common_elements(set1, set2):
    return set(set1) & set(set2)

seta = [1,2,3,4,5,6]
setb = [4,5,6,7,8,9]
print(common_elements(seta,setb))
```

```python
# Write a function lcm(a, b) that takes two integers and returns their least common multiple

import math

def lcm(a, b):
    if a == 0 or b == 0:
        return 0 
    return abs(a * b) // math.gcd(a, b)

print(lcm(12, 18))  
print(lcm(5, 7))   
```

```python
#qn 2
def sum_natural(n):
    if n == 0:
        return 0
    return n + sum_natural(n-1)
```

```python
#6
def is_palindrome(s):
    if len(s) == 0 or len(s) == 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])
```

```python
print(is_palindrome("binib"))
```

```python
#7
def reverse_string(s):
    if len(s) == 1:
        return s
    return s[-1] + reverse_string(s[0:-1])
```

```python
#tower of hanoi

def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    
    # move n-1 disks to auxiliary
    tower_of_hanoi(n-1, source, destination, auxiliary)
    
    # move largest disk
    print(f"Move disk {n} from {source} to {destination}")
    
    # move n-1 disks to destination
    tower_of_hanoi(n-1, auxiliary, source, destination)

n = 3
tower_of_hanoi(n, 'A', 'B', 'C')
```

```python
#alternate

def hanoi(n, s, a, d):
    if n:
        hanoi(n-1, s, d, a)
        print(f"{s} â†’ {d}")
        hanoi(n-1, a, s, d)
        
n = 3
tower_of_hanoi(n, 'A', 'B', 'C')
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __iadd__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __mul__(self, scalar):
        return Point(self.x * scalar, self.y * scalar)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __lt__(self, other):
        return self.distance_from_origin() < other.distance_from_origin()

    def __gt__(self, other):
        return self.distance_from_origin() > other.distance_from_origin()
    
    def __hash__(self):
        return hash((self.x,self.y))
    
    def distance_from_origin(self):
        return (self.x**2 + self.y**2)**(1/2)
    
    def __str__(self):
        return f"({self.x},{self.y})"
    
    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"
```

```python
p1 = Point(2,2)
p2 = Point(5,2)
print(p1 + p2)
print(p1 * 2)
print(p1 == p2)
print(p1 > p2)
print(p1 < p2)
p3 = Point(1,2)
p3 += p2   #iadd
print(p3)
```

```python
#using map create two functions for encrypting and decrypting message using ceaser cypher
```

```python
# function to shift a single character
def shift_char(c, shift):
    if c.isalpha():  # only letters
        base = ord('A') if c.isupper() else ord('a')
        return chr((ord(c) - base + shift) % 26 + base)
    return c  # keep spaces/symbols unchanged

# encrypt function
def encrypt(message, shift):
    return "".join(map(lambda c: shift_char(c, shift), message))

# decrypt function
def decrypt(cipher, shift):
    return "".join(map(lambda c: shift_char(c, -shift), cipher))

msg = "Im Binsa Hii"
shift = 3

enc = encrypt(msg, shift)
dec = decrypt(enc, shift)

print("Encrypted:", enc)
print("Decrypted:", dec)
```

```python
# 10 floor tall building. an egg when dropped breaks from certain floor. 
# Value for each batch of egg is different and random
# find floor where egg breaks from sequential and binary search
```

```python
#sequential search
import random
break_floor = random.randint(1,10)
for floor in range(1,11):
    if floor >= break_floor:
        print("egg breaks in floor",floor)
        break
```

```python
target = random.randint(1,10)
low = 0
high = 10
while low < high:
    mid = (high + low)//2
    if mid == target:
        print("egg breaks in floor",mid)
        break
    elif target < mid:
        high = mid
    elif target > mid :
        low = mid
```

```python
# do using list
floors = list(range(11))
target = 5
l, h = 0, 10
while l <= h:
    mid = (l+h)//2
    if floors[mid] == target:
        print("Egg breaks on floor", floors[mid])
        break
    elif floors[mid] > target:
        h = mid
    else:
        l = mid
```
