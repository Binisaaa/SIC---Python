```python
#8-1
primes = []
composites = []
for num in range(2, 13):
    count = 0
    for i in range(1, num):
        if num % i == 0:
            count += 1
    if count == 1:
        primes.append(num)
    else:
        composites.append(num)
print("prime numbers are ",primes)
print("composite numbers are ",composites)
```

```python
#9 - 1
print("Armstrong numbers from 100 to 999: ")
for num in range (100,1000):
    a = num // 100 
    b = (num // 10) % 10
    c = num % 10 
    if a**3 + b**3 + c**3 == num:
        print(num)
```

```python
#12-01 using tuple data ,2 variables can be swapped,using this swap largest value of 12 with last, use for loop
lst = [5, 6, 3, 9, 2, 12, 3, 8, 7]
largest = max(lst)
last_index = len(lst) - 1

for i in range(len(lst)):
    if lst[i] == largest:
        lst[i], lst[last_index] = lst[last_index], lst[i] 
        break
print(lst)
```
